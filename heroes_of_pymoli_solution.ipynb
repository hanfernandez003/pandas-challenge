```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head(3)

```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].unique())
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item Name"].unique())
average_price  = '${:,.2f}'.format(purchase_data["Price"].mean())
total_purchases = len(purchase_data["Purchase ID"].unique())
total_revenue = '${:,.2f}'.format(purchase_data["Price"].sum())

purchase_analysis = pd.DataFrame({"Number of Unique Items": [unique_items],
                                   "Average Price": [average_price],
                                   "Number of Purchases": [total_purchases],
                                   "Total Revenue": [total_revenue]})

purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_data["Total Count"] = 1
gender_grouped = purchase_data.groupby('Gender')['Total Count'].count().reset_index()
gender_grouped ["Percentage of Players"] = gender_grouped['Total Count'] / gender_grouped['Total Count'].sum()

gender_grouped ["Percentage of Players"] = gender_grouped ["Percentage of Players"].map("{:.2%}".format)

gender_grouped.sort_values(["Total Count"], ascending = False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Price 2"] =purchase_data["Price"]

purchase_data_group = (purchase_data.groupby('Gender', as_index=False)
                      .agg({'Total Count':'sum','Price':'mean','Price 2':'sum'})
                      .rename(columns={'Total Count':'Purchase Count','Price':'Average Purchase Price', 'Price 2':'Total Purchase Value'}))

purchase_data_group["Average Purchase Price"] = purchase_data_group["Average Purchase Price"].map("${:,.2f}".format)
purchase_data_group["Total Purchase Value"] = purchase_data_group["Total Purchase Value"].map("${:,.2f}".format)

purchase_data_group
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 10, 14,19,24,29,34,39,45]
group_labels = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["Age Group"]= pd.cut(purchase_data["Age"], bins, labels=group_labels)


age_grouped = purchase_data.groupby('Age Group')['Total Count'].count().reset_index()
age_grouped ["Percentage of Players"] = age_grouped['Total Count'] / age_grouped['Total Count'].sum()

age_grouped ["Percentage of Players"] = age_grouped ["Percentage of Players"].map("{:.2%}".format)

age_grouped
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

purchase_age_group = (purchase_data.groupby('Age Group', as_index=False)
                      .agg({'Total Count':'sum','Price':'mean','Price 2':'sum'})
                      .rename(columns={'Total Count':'Purchase Count','Price':'Average Purchase Price', 'Price 2':'Total Purchase Value'}))

purchase_age_group["Average Purchase Price"] = purchase_age_group["Average Purchase Price"].map("${:,.2f}".format)
purchase_age_group["Total Purchase Value"] = purchase_age_group["Total Purchase Value"].map("${:,.2f}".format)

purchase_age_group
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender_group = (purchase_data.groupby('SN')
                      .agg({'Total Count':'sum','Price':'mean','Price 2':'sum'})
                      .rename(columns={'Total Count':'Purchase Count','Price':'Average Purchase Price', 'Price 2':'Total Purchase Value'})
                      .sort_values(["Total Purchase Value"], ascending=False))

top_spender_group["Average Purchase Price"] = top_spender_group["Average Purchase Price"].map("${:,.2f}".format)
top_spender_group["Total Purchase Value"] = top_spender_group["Total Purchase Value"].map("${:,.2f}".format)

top_spender_group.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_group = (purchase_data.groupby(['Item ID','Item Name'], as_index=False)
                      .agg({'Total Count':'sum','Price 2':'sum'})
                      .rename(columns={'Total Count':'Purchase Count', 'Price':'Item Price','Price 2':'Total Purchase Value'})
                      .sort_values(["Total Purchase Value"], ascending=False))
 
items_group["Total Purchase Value"] = items_group["Total Purchase Value"].map("${:,.2f}".format)    


items_group.head(5)
                                     
```
